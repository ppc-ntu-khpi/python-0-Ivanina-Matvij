```python
from math import sqrt,pi
S = float(input('Введіть площу кола: '))
r = sqrt(S/pi)
print ('Радіус дорівнює %6.3f см' %r)
```

```python
R = float(input("введіть радіус сфери = "))
S = 4 * pi * R**2
print ("Площа сфери = ",S)

R = float(input("введіть радіус кулі = "))
V = (4/3) * pi * R**3
print ("Об'єм кулі = ",V)
```
